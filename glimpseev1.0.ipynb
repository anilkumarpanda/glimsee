```python
#Add blob from pathm
from azure.storage.blob import BlockBlobService
from os import listdir
from os.path import isfile, join
import os

# Set parameters here
ACCOUNT_NAME = "glimpsee"
ACCOUNT_KEY = ""
CONTAINER_NAME = "afws"
LOCAL_DIRECT = "C:\\Users\\Owner\\Pictures\\4K Stogram\\awakenings"        

blob_service = BlockBlobService(account_name=ACCOUNT_NAME, account_key=ACCOUNT_KEY)
# find all files in the LOCAL_DIRECT (excluding directory)
local_file_list = [f for f in listdir(LOCAL_DIRECT) if isfile(join(LOCAL_DIRECT, f))]
included_extenstions = ['jpg','jpeg']
local_file_list = [fn for fn in local_file_list if any(fn.endswith(ext) for ext in included_extenstions)]
file_num = len(local_file_list)
counter = 0
for i in range(file_num):
    local_file = join(LOCAL_DIRECT, local_file_list[i])
    blob_name = local_file_list[i]
    try:
        #blob_service.put_block_blob_from_path(CONTAINER_NAME, blob_name, local_file)
        blob_service.create_blob_from_path(CONTAINER_NAME,blob_name,local_file)
        print (" Success !! %s"%blob_name)
    except:
        print ("Something WRONG happened when uploading the data %s"%blob_name)
```
